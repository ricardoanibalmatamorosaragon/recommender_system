```python
from nltk.cluster.util import cosine_distance
from collections import Counter
import csv
import numpy as np
import itertools
import string
import networkx as nx
import re
import nltk
import matplotlib.pyplot as plt

nltk.download('wordnet')
#nltk.download('stopwords')
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
from nltk.tokenize import RegexpTokenizer
from nltk.stem.wordnet import WordNetLemmatizer

import pandas as pd
```

```python
dataset = pd.read_csv("dataset_risorse.csv",delimiter=",")
dataset.head(100)
```

```python
#esplorazione preliminare delle keywords
dataset['word_count'] = dataset['keywords'].apply(lambda x: len(str(x).split(" ")))
dataset[['keywords','word_count']].head()
```

```python
#statistiche descrittive del conteggio delle parole
#indicazione della dimensione del set di dati che 
#stiamo gestendo e della variazione del conteggio delle 
#parole tra le righe.
dataset.word_count.describe()
```

```python
total_keywords=''
for index,row in dataset.iterrows():
    temp=row['keywords'].replace("'",'')
    temp=temp.replace(',','')
    total_keywords+=' '+temp
#total_keywords
```

```python
#split() returns list of all the words in the string 
split_keywords = total_keywords.split()
# Pass the split_it list to instance of Counter class. 
count_k = Counter(split_keywords) 

  
# most_common() produces k frequently encountered 
# input values and their respective counts. 
most_occur = count_k.most_common(40)

data_plot = pd.DataFrame(columns=['keyword','count'])
keyws=[]
count=[]
for i in most_occur:
    keyws.append(i[0])
    count.append(i[1])
data_plot['keyword']=keyws
data_plot['count']=count
data_plot.plot(x ='keyword', y='count', kind = 'bar')
```

```python
most_occur[-10:]
```

```python
#Creating a list of stop words and adding custom stopwords
stop_words = stopwords.words("english","italian")
```

```python
corpus = []
for i in range(len(split_keywords)):
    #Remove punctuations
    text = re.sub('[^a-zA-Z]', ' ', split_keywords[i])
    
    #Convert to lowercase
    text = text.lower()
    
    #remove tags
    text=re.sub("&lt;/?.*?&gt;"," &lt;&gt; ",text)
    
    # remove special characters and digits
    text=re.sub("(\\d|\\W)+"," ",text)
    
    #Stemming
    ps=PorterStemmer()
    text=ps.stem(text)
    
    #Lemmatisation
    lem = WordNetLemmatizer()
    if not text in stop_words:
        text=lem.lemmatize(text) 
    
    corpus.append(text)
    
#stopwords
filtered_corpus = [word for word in corpus if word not in stop_words]
```

```python
#delete duplicated
corpus_without_duplicated= set(corpus)
print('keywords without duplicated :',str(len(corpus_without_duplicated)))
```

```python
#Word cloud
from os import path
from PIL import Image
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
import matplotlib.pyplot as plt
%matplotlib inline
wordcloud = WordCloud(
                          background_color='white',
                          stopwords=stop_words,
                          max_words=100,
                          max_font_size=50, 
                          random_state=42
                         ).generate(str(filtered_corpus))
print(wordcloud)
fig = plt.figure(1)
plt.imshow(wordcloud)
plt.axis('off')
plt.show()
fig.savefig("word1.png", dpi=900)
```
